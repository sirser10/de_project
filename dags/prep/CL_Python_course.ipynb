### Вопрос 2. Проработайте исторические данные 

```python
historical_data_df.fillna(historical_data_df.mean(), inplace=True)
display(historical_data_df)
q1, q3 = np.percentile(historical_data_df['Revenue'], [25,75])
iqr=q3 - q1 

lower_thershold= q1 - 1.3 * iqr
upper_thershold= q3 + 1.3 * iqr

historical_data_df['rolling_mean']=historical_data_df['Revenue'].rolling(window=7, min_periods=1, center=False).mean()

historical_data_df['processed_quartiles']=np.where(
    (historical_data_df['Revenue']<lower_thershold) | (historical_data_df['Revenue']>upper_thershold),
    historical_data_df['rolling_mean'].shift(1),
    historical_data_df['Revenue']
    )

historical_data_df.drop(columns=['rolling_mean'],inplace=True)
display(historical_data_df)

fix=px.line(historical_data_df, x='dates', y=['Revenue','processed_quartiles'], title='Revenue на каждый день', range_y=[0, 10000])
fix.show()
```
